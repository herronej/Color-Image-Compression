```python
# function for calculating Euclidean distance
def euc_dist(x, y):
    
    diff = x-y
    
    if(diff.ndim == 1):
        return np.abs(diff)
    
    return np.linalg.norm(diff, axis = 1)

def kmeans(X, k=256):
    prev_k_means = np.array([[0 for i in range(3)] for j in range(k)])
    means = np.array([[np.random.randint(INT_MIN, INT_MAX) for i in range(3)] for j in range(k)])
    indices = np.array([ind for ind in range(len(X))])
    #print('means', means)
    '''print(indices)'''
    
    count = 0
    nearest_means = []
    #prev_nearest_means = [0 for i in range(len(X))]
    while not np.array_equal(prev_k_means, means) and count < 1000: #for c in range(5):#while(prev_features != features):
        count += 1
        #print(count)
        nearest_means = []
        # find nearest means
        for x in X:
            s = euc_dist(means, x)
            #print('s', s)
            nearest_means.append(np.argmin(s))
        nearest_means = np.array(nearest_means)
        '''print('X', X)
        print('nearest_means', nearest_means)
        print('k=0', X[nearest_means[:]==0])'''
        prev_k_means = means
        means = np.array([[0,0,0] for i in range(k)])
        for k_ in range(k):
            cluster = X[nearest_means[:]==k_]
            if len(cluster) == 0:
                means[k_] = prev_k_means[k_]
            else:
                means[k_] = np.array([np.mean(cluster, axis=0)])
        '''print(np.mean(X[nearest_means[:]==104], axis=0))
        print(k_means.shape)'''
        #print(means)
    means = np.array([[int(np.round(means[j,i])) for i in range(3)] for j in range(k)])
    return nearest_means, means

def silhouette_score(X, nearest_means, means, k=256):
    
    scores = []
    for i in range(len(nearest_means)):
        # calculate A
        cluster = X[nearest_means[:]==nearest_means[i]]
        if len(cluster) == 0:
            a = 0.0
        else:
            a = np.mean(euc_dist(cluster, X[i]), axis = 0)

        # calculate B
        indices = [j for j in range(k)]
        dists = euc_dist(X[i], means)
        s = np.column_stack((indices, dists))
        s = s[s[:,1].argsort()]
        
        # get second nearest cluster
        cluster_ind = int(s[1,0])
        cluster = cluster = X[nearest_means[:]==nearest_means[cluster_ind]]
        if len(cluster) == 0:
            b = 0.0
        else:
            b = np.mean(euc_dist(cluster, X[i]), axis = 0)
        
        '''print('a', a)
        print('b', b)'''
        if max(a,b) == 0:
            score = 0.0
        else:
            score = (b-a)/max(a,b)
        scores.append(score)
    return np.mean(score)

def calculate_error(features, nearest_means, means):
    sum = 0.0
    for i in range(len(nearest_means)):
        sum += (np.linalg.norm(np.subtract(features[i], means[nearest_means[i]])))
    return sum/len(nearest_means)

scores = []
ks = [2,4,8,16,32,64,128,256]
for k_ in ks:
    nearest_means, means = kmeans(features, k=k_)
    print(nearest_means)
    print(means)
    scores.append(calculate_error(features, nearest_means, means))
    print(scores)#silhouette_score(features, nearest_means, means, k=k_))
```

```python
# plot silhouette scores
def plot_error(k, scores, title, xlabel, ylabel):
    
    plt.plot(k, scores)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.legend()
    plt.show()
    
plot_error(ks, scores, 'K-Means Silhouette Scores', 'K', 'Silhouette Score')
```

```python
def means_to_img(means, nearest_means):

    new_r = []
    new_g = []
    new_b = []

    for i in range(IMG_DIM):
        new_r_row = []
        new_g_row = []
        new_b_row = []
        for j in range(IMG_DIM):
            new_r_row.append(means[nearest_means[i*IMG_DIM+j],0])
            new_g_row.append(means[nearest_means[i*IMG_DIM+j],1]) 
            new_b_row.append(means[nearest_means[i*IMG_DIM+j],2])
        new_r.append(new_r_row)
        new_g.append(new_g_row)
        new_b.append(new_b_row)
     
    '''print(new_r)
    print(new_g)
    print(new_b)'''
    new_img = np.stack(np.array([new_r, new_g, new_b]), axis=2)
    print(new_img)
    plt.imshow(new_img)
    
means_to_img(means, nearest_means)
```

```python
def winner_take_all(X, epsilon=0.01, k=256):
    prev_nearest_means = np.array([0 for j in range(len(X))])
    nearest_means = np.array([1 for j in range(len(X))])
    means = np.array([[np.random.uniform(INT_MIN, INT_MAX) for i in range(3)] for j in range(k)])
    indices = np.array([ind for ind in range(k)])
    #print('means', means)
    print(indices)
    count = 0
    #nearest_means = []
    #prev_nearest_means = [0 for i in range(len(X))]
    while not np.array_equal(prev_nearest_means, nearest_means) and count < 1000:
        prev_nearest_means = np.copy(nearest_means)
        
        count += 1
        print(count)
        for i in range(len(X)):
            
            dists = euc_dist(means, X[i])
            s = np.column_stack((indices, dists))
            s = s[s[:,1].argsort()]
            
            # get nearest cluster
            cluster_ind = int(s[0,0])
            #print(cluster_ind)
            means[cluster_ind] = means[cluster_ind] + epsilon*(np.subtract(X[i], means[cluster_ind])) 
            nearest_means[i] = cluster_ind
            
    means = np.array([[int(np.round(means[j,i])) for i in range(3)] for j in range(k)])
    return np.array(nearest_means), means

nearest_means, means = winner_take_all(features, epsilon=0.01, k=256)
print(calculate_error(features, nearest_means, means))
```

```python
def mle_mean(data):
    n = len(data)
    mean = np.sum(data, axis=0)/n
    return mean

def mle_stddev(data):
    
    n = len(data)
    sample_mean = np.sum(data, axis=0)/n 
    diff = data - sample_mean
    square_diff = np.square(diff)
    col_sums = np.sum(square_diff, axis=0)
    variance = col_sums/n
    std_dev = np.sqrt(variance)
    
    return std_dev

def mle_covariance(data, mu):
    
    n = len(data)
    diff = data - mu
    diff = diff.T
    product = np.matmul(diff, diff.T)
    covariance = product/n
    return covariance


mle_stddev(features)
mle_covariance(features, mle_mean(features))
```

```python
def kernel(x, h):
    if np.linalg.norm(x) <= h:
        return 1
    else:
        return 0
    
def mean_shift(X, h):
    
    indices = np.array([int(ind) for ind in range(len(X))])
    print(indices)
    for i in range(0, len(X)):
            
        dists = euc_dist(X, X[i])
      
        s = np.column_stack((indices, dists))
        
        #sort by distance
        s = s[s[:,1].argsort()]
        s = s[:h,]
        
        num = 0.0
        denom = 0.0
        
        for j in range(len(s)):
            #print(s[j,0])
            '''print(X[i])'''
            num += kernel(X[int(s[j,0])]-X[i], h)*int(s[j,0])
            denom += kernel(X[int(s[j,0])]-X[i], h)
            
        # shift x
        X[i] = num/denom
    
    return X
        
X = mean_shift(features, 2)
```
